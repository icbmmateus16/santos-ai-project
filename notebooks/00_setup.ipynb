## 2. Verificar versões das bibliotecas

```python
import importlib

libs = [
    'pandas', 'numpy', 'scipy', 'matplotlib', 'seaborn', 'plotly',
    'sklearn', 'shap', 'requests', 'bs4', 'wikipediaapi', 'wikipedia',
    'nltk', 'textblob', 'wordcloud', 'missingno', 'tqdm'
]

print(f"{'Biblioteca':<20} {'Status':<10} {'Versão'}")
print('-' * 50)
for lib in libs:
    try:
        mod = importlib.import_module(lib)
        version = getattr(mod, '__version__', 'OK')
        print(f"{lib:<20} {'✓':<10} {version}")
    except ImportError:
        print(f"{lib:<20} {'✗ FALTA':<10} — instale com pip")
```

## 3. Criar estrutura de pastas

```python
import os
from pathlib import Path

# Detecta automaticamente a raiz do projeto
PROJECT_ROOT = Path(os.getcwd()).parent
print(f'Raiz do projeto: {PROJECT_ROOT}')

pastas = [
    'data/raw',
    'data/processed',
    'data/external',
    'notebooks',
    'outputs/figures',
    'outputs/models',
]

for pasta in pastas:
    caminho = PROJECT_ROOT / pasta
    caminho.mkdir(parents=True, exist_ok=True)
    print(f'  ✓ {pasta}')

print('\nEstrutura criada com sucesso!')
```
